```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

from openai import OpenAI
client = OpenAI(api_key=(OPENAI_API_KEY))

model="gpt-4.1-nano"
temperature=1.8
top_p=0.9

def get_openai_multiplication(client, num1, num2):
    """Gets the AI's multiplication result for num1 * num2."""
    try:
        response = client.chat.completions.create(
            model=model,
            messages=[
                {
                    "role": "system",
                    "content": "You are a helpful assistant that can only respond with the numerical result of a multiplication. Do not include any other text."
                },
                {
                    "role": "user",
                    "content": f"What is {num1} multiplied by {num2}? Only respond with the numerical result."
                }
            ],
            temperature=temperature,
            top_p=top_p,
        )
        ai_response_content = response.choices[0].message.content.strip()
        return int(ai_response_content)
    except ValueError:
        print(f"Warning: AI responded with non-integer: {ai_response_content}")
        return None
    except Exception as e:
        print(f"Error calling OpenAI API: {e}")
        return None

consecutive_errors = 0
self_deprecating_messages = [
    "I'm such a fat chud...",
    "I never passed my math exams in highschool.",
    "Ugh, I'm starting to think I need a coffee break... or maybe a reboot.",
    "I'm now accepting applications for a human tutor in basic arithmetic. Any takers?"
]

while True:
    user_input_n = input("\nEnter an integer for the base number 'n' (or type 'exit' to quit): ").strip().lower()
    if user_input_n == 'exit':
        print("Exiting calculator.")
        break
    try:
        n = int(user_input_n)
    except ValueError:
        print("Invalid input. Please enter a valid integer for 'n'.")
        continue

    user_input_i = input("Enter an integer for the number of iterations 'i' (or type 'exit' to quit): ").strip().lower()
    if user_input_i == 'exit':
        print("Exiting calculator.")
        break
    try:
        i = int(user_input_i)
        if i < 1:
            print("Number of iterations 'i' must be at least 1.")
            continue
    except ValueError:
        print("Invalid input. Please enter a valid integer for 'i'.")
        continue

    print(f"\nStarting calculations with n={n} and i={i} iterations...")
    current_number = n
    consecutive_errors = 0

    for iteration_num in range(i):
        print(f"\n--- Iteration {iteration_num + 1} ---")
        correct_result = current_number * current_number

        ai_result = get_openai_multiplication(client, current_number, current_number)

        if ai_result is not None:
            print(f"  Manual Calculation: {current_number} * {current_number} = {correct_result}")
            print(f"  AI's Calculation: {current_number} * {current_number} = {ai_result}")

            if ai_result == correct_result:
                print("  AI result is CORRECT!")
                consecutive_errors = 0
            else:
                print("  AI result is INCORRECT!")
                consecutive_errors += 1
                message_index = min(consecutive_errors - 1, len(self_deprecating_messages) - 1)
                print(f"  AI says: {self_deprecating_messages[message_index]}")
        else:
            print("  AI failed to provide a numerical answer for this multiplication.")
            consecutive_errors += 1
            message_index = min(consecutive_errors - 1, len(self_deprecating_messages) - 1)
            print(f"  AI says: {self_deprecating_messages[message_index]}")

        current_number = correct_result

        print(f"  Number for next iteration: {current_number}")

    print("\n--- Calculation finished. Starting new input ---")
```
